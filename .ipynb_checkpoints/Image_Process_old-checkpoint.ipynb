```python
# Helper functions for generate fake data and visualize

#Credit to Ariana's code
def mask_circle(shape,h,k,r):
    """
    Generate a 2d array of input shape, where pixels that lie within the circle of radius r and
    center (h,k) are set to have a value of 1.
    
    Inputs:
    -------
    shape: (x,y) dimensions of 2d array to generate
    h,k: x,y positions of circle's center #relative to the image, not the whole canvas!
    r: radius of circle
    
    Outputs:
    --------
    cmask: 2d array with pixels in specified circle given a value of 1
    """
    xx, yy = np.meshgrid(range(shape[1]), range(shape[0]))
    radii = np.sqrt(np.square(xx - h) + np.square(yy - k))
    cmask = np.zeros(shape, dtype=int)
    cmask[radii<r] = 1
    return cmask

'''
Generate tiles from a large tilt image (canvas).

Input:
------
canvas: numpy 2d array, region of interest of the sample.
beam_centers: numpy 2d array of shape (C, 2), where C is the number of beams/tiles.
radius: float, beam radius.
circle: bool, whether the beam is circle (0-mask the corners) or square.
        This parameter will be change to mask to receive arbitrary shape beam.
prep: bool, whether preprocess (zero-center normalize) each tile independently.

Output:
-------
Ims: numpy 3d array of shape (C, 2*radius, 2*radius), all tiles.
'''
def get_tiles(canvas, beam_centers, radius, circle=True, prep=True):
    upleft = np.array(np.floor(beam_centers - radius),int)
    dim = int(2*radius)
    if circle:
        mask = mask_circle((dim,dim), rad, rad, rad)
    else: mask = 1
    Ims = []
    for i in range(beam_centers.shape[0]):
        print(f'generating tile {i}')
        
        tile = ma.masked_array(canvas[upleft[i,0]:upleft[i,0]+dim,
                                      upleft[i,1]:upleft[i,1]+dim]*mask, 1-mask)
        Ims.append(tile)
    
#     if prep: Ims = [preprocess(tile) for tile in Ims] #each tile independently preprocess
    if prep: Ims = preprocess(Ims) #preprocess all tiles together
    Ims = ma.array(Ims)
    return Ims

'''
Convert a colorful image to gray image.
'''
def rgb2gray(rgb):
    return np.dot(rgb, [0.2989, 0.5870, 0.1140])
```

```python
'''
Cut a large canvas into series of smaller tiles.

Input:
------
ims: 2d numpy array, the larger canvas image
pixel_size: float, pixel size in unit of Å
bound_A: float, search boundary length in one direction in unit Å. 
         What matters is only the ratio of bound_A/pixel_size.
n_beams: int, number of tiles. The first n_beams centers from true_centers
         will be extracted. 
beam_d: int, diameter of each beam, in unit of pixels
circle: bool, whether mask each tile as a perfect circle beam. 
        This will be changed as the parameters for get_tiles() changed.
prep: bool, whether preprocess the tiles or not.

Output:
-------
Ims: numpy array of shape (C, beam_d, beam_d), all tiles.
true_centers: numpy array of shape (C,2), ground truth of tile centers.
centers0: numpy array of shape (C,2), initial guesses of tile centers.
pairs: list of tuples of indices of tiles that have overlapping regions.
bound_p: float, =bound_A/pixel_size.
'''
def simulate_canvas_tiles(ims, true_centers, pixel_size=2, bound_A=20,
                          n_beams=7, beam_d=400, circle=True, prep=False,
                          perturb_true=False):
    bound_p = bound_A/pixel_size
    rad = beam_d/2
    true_centers = beam_centers[:n_beams, :]*rad/(beam_diameter/2) + np.array(ims.shape)/2
    true_centers = np.floor(true_centers)
    if perturb_true: 
        print('Intended beam centers before shift error simulation')
        print(true_centers)
        true_centers += np.random.randint(-bound_p,bound_p+1,(n_beams,2))
    pairs = get_pairs(true_centers, beam_d)
    print(pairs)

    Ims = get_tiles(ims, true_centers, rad, circle=circle, prep=prep)
    if not perturb_true: 
        overlap = cal_overlap(1-Ims[0,:,:].mask, 1-Ims[1,:,:].mask, 
                              true_centers[0,:], true_centers[1,:],
                              purpose='fraction')
        print(f'overlap% {overlap}')
    else:
        for p in pairs:
            overlap = cal_overlap(1-Ims[p[0],:,:].mask, 1-Ims[p[1],:,:].mask, 
                              true_centers[p[0],:], true_centers[p[1],:],
                              purpose='fraction')
            print(f'overlap% {overlap} for tile ({p[0]}, {p[1]})')

    centers0 = true_centers.copy()
    centers0[1:,:] += np.random.randint(-bound_p,bound_p+1,centers0[1:,:].shape)
    return (Ims, true_centers, centers0, pairs, bound_p)


'''
Simulate the greedy optimization process with increasing random noise.

Input:
------
Is: numpy array of shape (C, beam_d, beam_d), all tiles.
initial_guesses:umpy array of shape (C,2), initial guesses of tile centers.
true_centers: numpy array of shape (C,2), ground truth of tile centers.
bound_p: int, +/- search boundaries from initial guesses in unit of pixels.
canvas_shape: tuple of ints, shape of the larger stitched image.
im_path: string, path to file that stores all numpy nd array tiles.
c_path: string, path to file that stores initial guesses of the centers.
pair_path: string, path to file that stores list of pairs.
mask: numpy array of shape (C, beam_d, beam_d), the mask used to generate
      numpy masked array of all tiles. That is, 0 means not masked means
      valid pixel in the beam region.
n_processes: int, number of processors for parallelization.
n_noise, n_step: int, random noise of standard deviation range(n_noise, nstep)
plot_stitch: bool, plot the stitched image of true and optimized centers for
             each noise introduced.
plot_CC: bool, plot the heatmap of normalized cross correlation matrices for
         the first pair of tiles. This makes sense when there is only one pair.
         
Output:
-------
times: 1d numpy array of length n_noise, storing the search time in seconds.
off_centers: numpy array of shape (n_noise, C, 2), storing the difference
             between optimized and true centers at each noise_std.
std2CC: dictionary from noise_std to a tuple of (cmat and nmat), where cmat
        is stores raw cross correlation scores and nmat stores the number of
        overlapping pixels to normalize the corresponding entry of raw cc.
        Both are numpy arrays of shape (C-1, P, 2*bound_p+1, 2*bound_p+1), 
        where C is the number of tiles (the first tile is fixed by default),
        P is the number of pairs involving the tile center to be searched 
        and already fixed tiles.
'''
def simulate_with_noise(Is, initial_guesses, true_centers, bound_p, canvas_shape,
                        im_path, pair_path, c_path, n_processes=4, 
                        n_noise=6, n_step=1, verbose=True, plot_stitch=True, 
                        plot_CC=False, early_stop=False):
    if verbose: 
        print(f'num processes:{n_processes}')
        print(f'search over ±{bound_p} bounds')
    mask = Is.mask
    Is = Is.data
    times = np.zeros(n_noise)
    off_centers = np.zeros((n_noise, initial_guesses.shape[0]-1,initial_guesses.shape[1]))
    std2CC = dict()
    
    for noise_std in range(0,n_noise,n_step):
        if verbose: print(f'introduce noise {noise_std}')
        Images = Is.copy()
        Images += np.random.normal(scale=noise_std, size=Images.shape)
        Images = ma.masked_array(Images, mask)
        Images = preprocess(Images)
        if verbose: print(f'finish preprocessing noisy tiles')
        np.save(c_path, initial_guesses)
        np.save(im_path, Images.data)       
        true_stitch = stitch(Images, true_centers, np.array([[0,0],canvas_shape]), 
                             plot=plot_stitch)

        start_time=time.time()
        opt_c, cmat, nmat = opt_CC_greedy(im_path, pair_path, c_path, bound_p, 
                                          debug=verbose, n_processes=n_processes)
        t = time.time() - start_time
        times[noise_std] = t
        if verbose: print(f'time: {t}s')
            
        std2CC[noise_std] = (cmat, nmat)
        my_stitch = stitch(Images, opt_c, np.array([[0,0],canvas_shape]), plot=False)
        #plot difference map
        if plot_stitch:
            plt.figure()
            plt.imshow(my_stitch-true_stitch, cmap='gray')
            plt.title(f'stitch difference map with noise {noise_std}')
            plt.clim(-2,2)
            plt.colorbar()
        #plot cc heatmap
        if plot_CC:
            plt.figure()
            norm_c = cmat[0][0]/nmat[0][0] #only plot the very first pair
            plt.imshow(norm_c, cmap='coolwarm', extent=[initial_guesses[1,1]-bound_p,
                                                        initial_guesses[1,1]+bound_p,
                                                        initial_guesses[1,0]+bound_p,
                                                        initial_guesses[1,0]-bound_p])
            plt.plot(true_centers[1,1], true_centers[1,0],'ko', markersize=3)
            plt.plot(initial_guesses[1,1], initial_guesses[1,0],'go', alpha=0.3,markersize=3)
            plt.plot(opt_c[1,1], opt_c[1,0],'go', markersize=3)
            plt.clim(-2,2)
            plt.title(f'normalized CC with noise {noise_std}')
            plt.colorbar()
            
        if np.array_equal(opt_c, true_centers): 
            print('Optimization succeeded.')
        else: 
            print(f'Optimization failed at noise_std {noise_std}.')
            off_centers[noise_std,:,:] = opt_c[1:,:] - true_centers[1:,:]
            if early_stop: return noise_std
    
    if early_stop: return n_noise+1
    return (times, off_centers, std2CC)
```

#### test different ordering for greedy search

```python
neighbors = dict()
for p in pairs:
    if p[0] in neighbors.keys(): neighbors[p[0]] += [p[1]]
    else: neighbors[p[0]] = [p[1]]
    if p[1] in neighbors.keys(): neighbors[p[1]] += [p[0]]
    else: neighbors[p[1]] = [p[0]]
neighbors
```

```python
def generate_one_order(neighbor_dict, num_b):
    order = [np.random.choice(num_b,1)[0]]
    while len(order) < num_b:
        density = np.ones(len(neighbor_dict[order[-1]]))
        for i in range(len(neighbor_dict[order[-1]])):
            if neighbor_dict[order[-1]][i] in order: density[i] = 0
        density /= sum(density)
        order.append(np.random.choice(neighbor_dict[order[-1]],size=1,p=density)[0])
    return order

def generate_orders(neighbor_dict, num_b, order):
    if len(order) == num_b: return [order]
    if len(order) == 0: 
        orders = []
        for n in range(num_b):
            orders += generate_orders(neighbor_dict, num_b, [n]) 
        return orders
    
    orders  = []
    n_set = []
    for fixed in order: n_set += neighbor_dict[fixed]
    n_set = list(set(n_set))
    for n in n_set:
        if n not in order: 
            orders += generate_orders(neighbor_dict, num_b, order+[n])
    return orders
```

```python
all_permutes = generate_orders(neighbors,nb,[])
```

```python
nrepeats = 3
max_n_noise = 5
n_sele = 100
selected = np.random.choice(len(all_permutes),n_sele,replace=False)
failed_indices = np.zeros(n_sele)
for i in range(len(selected)):
    order_ind = all_permutes[selected[i]]
    print(order_ind)

    Ims = Is[order_ind,:,:].copy()
    true_centers = ground_truth[order_ind,:].copy()
    centers0 = initial_guesses[order_ind,:].copy()
    new_pairs = get_pairs(true_centers, beam_d)
    np.save(pair_path, new_pairs)
    np.save(c_path, centers0)
    np.save(im_path, Ims.data)
    
    failed_ind = 0
    for repeat in range(nrepeats):
        start = time.time()
        first_failed_noise = simulate_with_noise(Ims, centers0, true_centers, bound_p, 
                                                 ims.shape, im_path, pair_path, c_path, 
                                                 n_processes=4, n_noise=max_n_noise, 
                                                 n_step=1, verbose=False, plot_stitch=False, 
                                                 plot_CC=False, early_stop=True)
        print(f'fail at {first_failed_noise} after {time.time()-start} s')
        failed_ind += first_failed_noise
    failed_indices[i] = failed_ind/nrepeats
```
